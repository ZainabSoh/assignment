```python
import pandas as pd

# Load the Titanic dataset

df = pd.read_csv('titanic.csv')
print(df)
```

```python
# Display the first few rows of the dataframe
print(df.head())

# Display the structure of the dataframe
print(df.info())

# Display statistical summary
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())

# Handle missing values: Fill 'Age' with the median, drop rows with missing 'Embarked'
df['Age'].fillna(df['Age'].median(), inplace=True)
df.dropna(subset=['Embarked'], inplace=True)

# Check for duplicates and remove them
df.drop_duplicates(inplace=True)

# Confirm that missing values are handled
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style
sns.set(style="whitegrid")

# Visualization 1: Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Visualization 2: Survived vs. Not Survived
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', data=df)
plt.title('Survival Counts')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()

# Visualization 3: Survival Rate by Class
plt.figure(figsize=(10, 6))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# Visualization 4: Survival Rate by Sex
plt.figure(figsize=(10, 6))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()

# Visualization 5: Age vs. Fare
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Age vs. Fare (Colored by Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```
